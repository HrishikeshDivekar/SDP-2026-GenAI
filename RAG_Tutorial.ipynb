```python
import requests
from langchain_core.documents import Document

# Always include https://
response = requests.get("https://www.themealdb.com/api/json/v1/1/search.php?f=a")

data = response.json()

# Extract the list of meals
meals = data.get("meals", [])

# Convert each meal into a Document
documents = [
    Document(page_content=meal["strMeal"], metadata=meal)
    for meal in meals
]

print(documents[0])
```
